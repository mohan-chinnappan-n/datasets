```python
import pandas as pd
```

```python
df1.head()
```

```python
df1
```

```python
df1.plot( grid=True, figsize=(30,20))
```

```python
df1[['Manufacturer','Sales_in_thousands' ]].plot(grid=True, figsize=(20,10))
# https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html
```

```python
df1[['Manufacturer','Sales_in_thousands' ]].iloc[1:30].plot.scatter(x="Manufacturer", y="Sales_in_thousands",   grid=True)
```

```python
df1[['Manufacturer','Sales_in_thousands' ]].iloc[1:20].plot.area(x="Manufacturer", y="Sales_in_thousands",   grid=True)
```
